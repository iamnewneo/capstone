```python
import glob
```

```python
all_images = glob.glob("./data/train/*.jpg")
```

```python
count = 0
total = len(all_images)
resolution_list_x = []
resolution_list_y = []
for img_path in all_images:
    if count%2000 == 0:
        print("Current: " +str(count+1)+"/"+str(total))
    img = Image.open(img_path).convert('RGB')
    width, height = img.size
    resolution_list_x.append(width)
    resolution_list_y.append(height)
    count+=1
```

```python
plt.scatter(resolution_list_x, resolution_list_y)
plt.xlabel("Image Width")
plt.ylabel("Image Height")
plt.title("Scatter plot of Resolution (Width vs Height) of training images.")
plt.savefig("scatter_plot_res.jpg")
plt.show()
```
